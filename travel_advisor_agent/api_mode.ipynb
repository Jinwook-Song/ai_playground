```python
import requests

BASE_URL = "http://127.0.0.1:8000"
APP_NAME = "travel_advisor_agent"
USER_ID = "u_123"
SESSION_ID = "29a3b2b3-6f91-4a4f-a79e-f662a27509af"
```

```python
response = requests.post(f"{BASE_URL}/apps/{APP_NAME}/users/{USER_ID}/sessions")

print(response.json())
```

```python
import sseclient

message = {
    "appName": APP_NAME,
    "userId": USER_ID,
    "sessionId": SESSION_ID,
    "newMessage": {"parts": [{"text": "일본 유후인 가고싶어"}], "role": "user"},
}

response = requests.post(
    f"{BASE_URL}/run_sse",
    json=message,
    stream=True,
)

client = sseclient.SSEClient(response)

for event in client.events():
    data = event.data
    content = data.get("content")
    parts = content.get("parts")
    for part in parts:
        function_call = part.get("functionCall", None)
        if function_call:
            print(function_call.get("name"))
        text = part.get("text", None)
        if text:
            print(text)
    print("=" * 60)
```
