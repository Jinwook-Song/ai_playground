```python
import requests

BASE_URL = "http://127.0.0.1:8000"
APP_NAME = "travel-advisor-agent"
USER_ID = "u_123"
SESSION_ID = "a555d066-c526-43dc-8f54-5336546f5ba3"
```

```python
response = requests.post(f"{BASE_URL}/apps/{APP_NAME}/users/{USER_ID}/sessions")

print(response.json())
```

```python
message = {
    "appName": APP_NAME,
    "userId": USER_ID,
    "sessionId": SESSION_ID,
    "newMessage": {
        "parts": [
            {"text": "일본 유후인 가고싶어"}
        ],
        "role": "user"
    } 
}

response = requests.post(f"{BASE_URL}/run", json=message)

# HTTP 상태 코드 확인
print(f"Status Code: {response.status_code}")

# 실제 응답 내용 확인 (JSON 파싱 전)
print(f"Response Text: {response.text}")

# Content-Type 확인
print(f"Content-Type: {response.headers.get('Content-Type')}")

# JSON 파싱 시도
try:
    print(response.json())
except Exception as e:
    print(f"JSON 파싱 실패: {e}")
```
